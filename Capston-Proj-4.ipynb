```python
movies = movies.fillna('')
```

```python
movies.head(1)
```

```python
movies['features'] = movies['genres'] #We will recommend all movies based on the genres
```

```python
#Now we will convert the texts into numbers
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(movies['features'])
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
def recommend_movie(movie_name):

    movie_name = movie_name.lower()

    movies['title_lower'] = movies['title'].str.lower()

    if movie_name not in movies['title_lower'].values:
        print("Movie not found")
        return

    index = movies[movies['title_lower'] == movie_name].index[0]

    similarity_scores = list(enumerate(similarity[index]))

    sorted_movies = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    print("Recommended Movies:\n")

    count = 1

    for movie in sorted_movies[1:11]:
        movie_index = movie[0]
        print(count, ".", movies.iloc[movie_index]['title'])
        count += 1
```

```python
recommend_movie("Toy Story (1995)")
```

```python
recommend_movie("Jumanji (1995)")
```

```python
recommend_movie("Jurassic Park")
```
